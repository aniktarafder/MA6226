```python
import numpy
import math
import vortex
from matplotlib import pyplot
%matplotlib inline
```

```python
N= 200
x_start, x_end = (-4, 4)
y_start, y_end = (-1, 1)
x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)
X, Y = numpy.meshgrid(x,y)
```

```python
#vortex
Nv = 10
xv_start, xv_end = -6, 6
yv_start, yv_end = 0, 0
xv = numpy.linspace(xv_start,xv_end, Nv)
yv = numpy.zeros_like(xv)
strength=1.0

u ,v = 0.0, 0.0


for i in range (Nv):
    
    u_vor, v_vor = vortex.vor(strength, xv[i], yv[i], X, Y)
    u = u + u_vor
    v = v + v_vor
    

xv
```

```python
width = 20
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)

pyplot.scatter(xv, yv, color='#CD2305', s=80, marker ='o');
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
```

```python
#analytical so;ution

a =( xv_end-xv_start )/(Nv-1)

bx = 2*math.pi*X / a
by = 2*math.pi*Y / a

ua =   strength / (2*a) * numpy.sinh(by) / (numpy.cosh(by)-numpy.cos(by)) 
va = - strength / (2*a) * numpy.sin(bx) / (numpy.cosh(by)-numpy.cos(bx)) 
```

```python
width = 10
height = (y_end - y_start) / (x_end - x_start) * width
pyplot.figure(figsize=(width, height))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)

pyplot.scatter
pyplot.streamplot(X, Y, ua, va)
```
